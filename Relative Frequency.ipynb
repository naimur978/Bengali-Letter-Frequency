```python
alphabets = ["অ","আ","ই","ঈ","উ","ঊ","ঋ","এ","ঐ","ও","ঔ","ক","খ","গ","ঘ","ঙ","চ","ছ","জ","ঝ","ঞ","ট","ঠ","ড","ঢ","ণ","ত","থ","দ","ধ","ন","প","ফ","ব",
"ভ","ম","য","র","ল","শ","ষ","স","হ","ড়","ঢ়","য়","ৎ","ং","ঃ","ঁ","া","ি","ী","ু","ূ","ৃ","ে","ৈ","ো","ৌ"]

freq = {}

def read_file(name = 'opohoron_novel.txt'):
    cou = 0
    with open(name, encoding="utf-8", errors='ignore') as f:
        for word in f:
            #print(word)
            for char in word:
                if char in freq:
                    freq[char]=freq[char]+1
                else:
                    freq[char]=int(1)
```

```python
def frequency(strings, letters):
    list_letter = []
    list_freq = []
    list_letter_freq = [list_letter,list_freq]
    for letter in letters:
        freq = 0
        for i in strings:
            if i == letter:
                freq += 1   
        list_letter.append(letter)
        list_freq.append(freq)
    return list_letter_freq 



def get_relative_freq(list_freq):
    list_rel_freq = []
    for i in list_freq:
        list_rel_freq.append((i/sum(list_freq))*100)
    return list_rel_freq
```

```python
with open('D:\Interim Files\Compressed\Bangla-Relative-Frequency\opohoron_novel.txt','r',encoding='utf-8') as f:
    string1 = f.read()
    


result = frequency(string1,alphabets)

mean_data = np.array(result)
dataset = pd.DataFrame({'Alphabet': mean_data[0,:], 'Count': mean_data[1,:]})


dataset['Count'] = dataset['Count'].astype('int')
dataset.sort_values(by='Count', ascending=False)
```
